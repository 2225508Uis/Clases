```python
import networkx as nx
import matplotlib.pyplot as plt
```

```python
class Grafo:
  
  def __init__(self,Qdirigido=False) -> None:
      if Qdirigido == True:
        self.G=nx.DiGraph()
        print("Se instancio un grafo dirigido")
      else:
        self.G=nx.Graph()   
        self.dirigido=True
        print("se instancio un grafo no dirigido")

  def addEdge(self,Edge, D=False ):
    if Edge == "":
      return
    w=0
    u,w,v=Edge.split()
    w=float(w)
    
    self.G.add_edge(u, v, weight=w)
    if D != True:
      self.G.add_edge(v, u, weight=w) 

  def printG(self,titulo):
    pos = nx.layout.spring_layout(self.G)
    nx.draw_networkx(self.G, pos)
    labels = nx.get_edge_attributes(self.G, "weight")
    nx.draw_networkx_edge_labels(self.G, pos, edge_labels=labels)
    plt.title(titulo)
    plt.show()
    
  def dkstra(self,start,end):
    path = nx.dijkstra_path(self.G, source=start, target=end)
    distance = nx.dijkstra_path_length(self.G, source=start, target=end)
    print("Ruta Por dikstra = ".format(start, end), path," La distancia es = ".format(start, end), distance)
    
  def dfs(self,start):
    ruta = list(nx.bfs_edges(self.G, source=start))    
    print("dfs es:",ruta,)
    
    
  def bfs(self, start):
    ruta = list(nx.dfs_edges(self.G, source=start))
    print("bfs es:",ruta)
    
```

```python
estaciones = Grafo(True)

estaciones.addEdge("a 5 b",True)
estaciones.addEdge("a 10 c",True)
estaciones.addEdge("b 15 d",True)
estaciones.addEdge("c 20 d",True)
estaciones.addEdge("c 5 e",True)
estaciones.addEdge("d 10 f",True)
estaciones.addEdge("e 15 f",True)
estaciones.addEdge("e 5 g",True)
estaciones.addEdge("f 10 g",True)
```

```python
#prueva de salida
estaciones.dfs("a")
estaciones.bfs("a")
estaciones.dkstra("a","g")
#prueva de salida
```
