```python
import matplotlib.pyplot as plt
import networkx as nx
```

```python
Grafo1=nx.Graph()
Grafo1.add_node("A")
Grafo1.add_node("B")
Grafo1.add_node("C")
Grafo1.add_node("D")
Grafo1.add_node("E")



Grafo1.add_edge("A", "B")
Grafo1.add_edge("A", "C")

Grafo1.add_edge("C", "B")
Grafo1.add_edge("C", "D")
Grafo1.add_edge("C", "E")

Grafo1.add_edge("B", "D")

Grafo1.add_edge("D", "E")



nx.draw(Grafo1, with_labels=True)
plt.show()
```

```python
Grafo2=nx.Graph()
Grafo2.add_nodes_from(["A","B","C","D","E"])
Grafo2.add_edges_from([("A", "B", weight=4), ("A", "C", weight=4), ("C", "B", weight=4), ("C", "D", weight=4), ("C", "E", weight=4), ("B", "D", weight=4), ("D", "E",weight=4)])


nx.draw(Grafo2, nx.spring_layout(Grafo2), with_labels=True,node_color="lightblue")
plt.show()
```
