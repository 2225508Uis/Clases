```python
import networkx as nx
import matplotlib.pyplot as plt
```

```python
def addEdge(grafo, u, v, w="", D=False):
  grafo.add_edge(u, v, weight=w)
  if D != True:
    grafo.add_edge(v, u, weight=w)
  pass


def printGraph(grafo):
  pos = nx.layout.spring_layout(grafo)
  nx.draw_networkx(grafo, pos)
  labels = nx.get_edge_attributes(grafo, "weight")
  nx.draw_networkx_edge_labels(grafo, pos, edge_labels=labels)
  plt.title("Grafo2")
  plt.show()
  pass


```

```python
grafo2 = nx.Graph()

addEdge(grafo2, "a", "b", 2)
addEdge(grafo2, "a", "h", 7)
addEdge(grafo2, "a", "g", 10)
addEdge(grafo2, "a", "b",2)
addEdge(grafo2, "a", "h",7)
addEdge(grafo2, "a", "g",4)

addEdge(grafo2, "b", "a",2)
addEdge(grafo2, "b", "g",2)
addEdge(grafo2, "b", "c",7)
addEdge(grafo2, "b", "d",2)
addEdge(grafo2, "c", "d",5)
addEdge(grafo2, "c", "b",7)
addEdge(grafo2, "c", "h",2)

addEdge(grafo2, "d", "b",2)
addEdge(grafo2, "d", "a",5)
addEdge(grafo2, "d", "e",1)
addEdge(grafo2, "d", "c",5)

addEdge(grafo2, "e", "d",1)
addEdge(grafo2, "e", "h",1)
addEdge(grafo2, "e", "g",3)
addEdge(grafo2, "e", "f",3)

addEdge(grafo2, "f", "e",3)
addEdge(grafo2, "f", "h",6)
addEdge(grafo2, "f", "g",3)

addEdge(grafo2, "g", "a",4)
addEdge(grafo2, "g", "b",2)
addEdge(grafo2, "g", "e",3)
addEdge(grafo2, "g", "f",3)

addEdge(grafo2, "h", "a",7)
addEdge(grafo2, "h", "f",6)
addEdge(grafo2, "h", "e",1)
addEdge(grafo2, "h", "c",2)

""" print(list(nx.bfs_edges(grafo2, source="a")))
print(list(nx.dfs_edges(grafo2, source="b"))) """

#Algoritmo dijkstra
start,end = input("Ingrese los nodos de inicio y fin separados por espacio = ").split()
path=nx.dijkstra_path(grafo2,source=start,target=end)
print("Ruta = ".format(start,end),path)
distance=nx.dijkstra_path_length(grafo2,source=start,target=end)
print("la distancia es = ".format(start,end),distance)

printGraph(grafo2)
```
