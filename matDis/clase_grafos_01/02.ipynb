```python
import networkx as nx
import matplotlib.pyplot as plt
```

```python
grafo1 = nx.Graph()

grafo1.add_edge("a","b",weight=10)
```

```python
nx.draw(grafo1, with_labels=True)
plt.show()
```

```python
def addEdge(grafo,u,v,w="",D=False):
  grafo.add_edge(u,v,weight=w)
  if D!=True :
    grafo.add_edge(v, u, weight=w)
  pass
def printGraph(grafo):
  pos = nx.layout.spring_layout(grafo)
  nx.draw_networkx(grafo,pos)
  labels=nx.get_edge_attributes(grafo,"weight")
  nx.draw_networkx_edge_labels(grafo,pos,edge_labels=labels)
  plt.title("Grafo")
  plt.show()
  pass
```

```python
grafo2 = nx.Graph()
addEdge(grafo2,"a","b",10)
printGraph(grafo2)
```
